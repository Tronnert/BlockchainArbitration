```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('time_output/part-00000.csv', delimiter='\t')
```

```python
def draw_avg_arb(data, asc=False, msg=""):
    sortd = data.sort_values('avg_arb', ascending=asc)[:10]

    fig = plt.figure(figsize=(5, 5))
    plt.xticks(rotation=90)
    plt.ylim (0, max(sortd['avg_arb']) + max(sortd['avg_arb']) * 0.1)
    plt.ylabel('Average duration')
    plt.xlabel('Currency pair - exchange pair')
    plt.title(msg)


    plt.bar(list(map(lambda x: '-'.join(x), sortd.loc[:, ('base', 'quote', 'bidExchange', 'askExchange')].values)), sortd['avg_arb'])
```

```python
draw_avg_arb(data, False, "Top-10 currency pairs by exchange pairs with longest average duration\n\n")
```

```python
draw_avg_arb(data, True, "Top-10 currency pairs by exchange pairs with shortest average duration\n\n")
```
